```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize']=10,6
```

```python
dataset = pd.read_csv("AirQualityUCI.csv", sep=";", decimal=",") 
dataset
```

```python
new = dataset.filter(['Date','RH'], axis=1)
```

```python


#new['Date'] = pd.to_datetime(new['Date'].astype(str), format='%d/%m/%Y')
#new['Date']
new_index = new.set_index(['Date'])
#new_index['Date'] = pd.to_datetime(new_index['Date'], infer_datetime_format=True) 
```

```python
new_index
```

```python
new_index.dropna(axis=0, how= 'all', inplace=True)
new_index
```

```python
from datetime import datetime
new_index.tail(5)
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.plot(new_index)
```

```python
rolmean = new_index.rolling(window=12).mean()
rolstd = new_index.rolling(window=12).std()
print(rolmean,rolstd)
```

```python
#plotrollingstatistics
orig = plt.plot(new_index,color='blue',label='original')
mean = plt.plot(rolmean,color='red',label='rolling mean')
std = plt.plot(rolstd,color='black',label='rolling std')
plt.legend(loc='best')
plt.title("rolling mean & standard deviation")
plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import adfuller
#dickyfuller
print('results of dickey-fuller test: ')
dftest = adfuller(new_index['RH'],autolag='AIC')

dfoutput = pd.Series(dftest[0:4],index=['Test Statistic','p-value','#Lags used','Number of observations used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key]=value
    
print(dfoutput) #Test Statistic < critical value then it h0: non-stationary fails and model is stationary
```

```python
new_index_logscale = np.log(new_index)
plt.plot(new_index_logscale)
```

```python
movingavg = new_index_logscale.rolling(window=12).mean()
movingstd = new_index_logscale.rolling(window=12).std()
plt.plot(new_index_logscale)
plt.plot(movingavg,color='red')
```

```python
datasetlogscaleminusmovingavg = new_index_logscale = movingavg
datasetlogscaleminusmovingavg.head(12)
datasetlogscaleminusmovingavg.dropna(inplace=True)
datasetlogscaleminusmovingavg.head(10)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationary(timeseries):
    
    movingavg = new_index_logscale.rolling(window=12).mean()
    movingstd = new_index_logscale.rolling(window=12).std()
    
    #plotrollingstatistics
    orig = plt.plot(new_index,color='blue',label='original')
    mean = plt.plot(rolmean,color='red',label='rolling mean')
    std = plt.plot(rolstd,color='black',label='rolling std')
    plt.legend(loc='best')
    plt.title("rolling mean & standard deviation")
    plt.show(block=False)
    #dickyfuller
    print('results of dickey-fuller test for RH : ')
    dftest = adfuller(new_index['RH'],autolag='AIC')

    dfoutput = pd.Series(dftest[0:4],index=['Test Statistic','p-value','#Lags used','Number of observations used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key]=value
    
    print(dfoutput)
```

```python
test_stationary(datasetlogscaleminusmovingavg)
```

```python
exponentialdecaywtavg = new_index_logscale.ewm(halflife=12,min_periods=0,adjust =True).mean()
plt.plot(new_index_logscale)
plt.plot(exponentialdecaywtavg, color='red')
```

```python
datasetlogscaleminusexponentialdecayavg = new_index_logscale - exponentialdecaywtavg
test_stationary(datasetlogscaleminusexponentialdecayavg)
```

```python
datasetlogshifting = new_index_logscale - new_index_logscale.shift()
plt.plot(datasetlogshifting)
```

```python
datasetlogshifting.dropna(inplace=True)
test_stationary(datasetlogshifting)
```

```python
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(new_index_logscale)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(new_index_logscale, label = 'original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='trend')
plt.legent(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='seasonal')
plt.legent(loc='best')
plt.subplot(414)
plt.plot(residual, label='residual')
plt.legent(loc='best')
plt.tight_layout()

decomposedlogdata = residual
decomposedlogdata.dropna(inplace=True)
test_stationary(decomposedlogdata)
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(new_index_logscale, order=(2,1,0))
results_AR = model.fit(disp=-1)
plt.plot(datasetlogshifting)
plt.plot(results_AR.fittedvalues,color='red')
plt.title('RSS %.4f'% sum(results_AR.fittedvalues-datasetlogshifting['RH'])**2)
print('plotting AR model')
```

```python
#MA model
model = ARIMA(new_index_logscale,order=(0,1,2))
results_MA = model.fit(disp=-1)
plt.plot(datasetlogshifting)
plt.plot(results_MA.fittedvalues, color = 'red')
plt.title('RSS %.4f'% sum(results_MA.fittedvalues-datasetlogshifting['RH'])**2)
print('plotting MA model')
```

```python
#ARIMA
model = ARIMA(new_index_logscale,order=(3,1,4))
results_ARIMA = model.fit(disp=-1)
plt.plot(datasetlogshifting)
plt.plot(results_ARIMA.fittedvalues, color = 'red')
plt.title('RSS %.4f'% sum(results_ARIMA.fittedvalues-datasetlogshifting['RH'])**2)
print('plotting ARIMA model')
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues,copy=True)
print(predictions_ARIMA_diff)
```

```python
prediction_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print(prediction_ARIMA_diff_cumsum)
```

```python
predictions_ARIMA_log = pd.Series(new_index_logscale['RH'].ix[0],index=new_index_logscale.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(prediction_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

```python
prediction_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(new_index)
plt.plot(prediction_ARIMA)
```
